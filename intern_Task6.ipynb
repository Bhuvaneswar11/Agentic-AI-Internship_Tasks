```python
def parking_lot_system(capacity,vehicle_logs):
    current_parked=0
    peak_usage=0

    for log in vehicle_logs:
        if log == "IN":
            current_parked+=1
            if current_parked > peak_usage:
                peak_usage=current_parked
        elif log == "OUT":
            if current_parked > 0:
                current_parked -=1

    print("Currently Parked vehicles:", current_parked)
    # print("Peak Parking Usage:", peak_usage)

    if current_parked > capacity:
        print("Parking Status: Exceeded Capacity! Alert!")
    else:
        print("Parking Status: Available")

capacity=50
vehicle_logs=["IN","IN","IN","OUT","IN","IN","OUT"]

parking_lot_system(capacity,vehicle_logs)
    
```

```python
def estimate_dlvry_time(distance,traffic,weather):
    base_speed=40

    base_time=(distance/base_speed)*60

    if traffic == "Low":
        base_time += 5
    elif traffic == "Medium":
        base_time += 10
    elif traffic == "High":
        base_time += 20

    if weather == "Rainy":
        base_time+=15
    elif weather == "Storm":
        base_time +=25

    print("Estimated Delivery Time:", round(base_time), "minutes")


distance=8
traffic="High"
weather="Rainy"

estimate_dlvry_time(distance, traffic, weather)
```

```python
def seat_booking(total_seats,booked_seats):
    booked_count=len(booked_seats)

    occupancy=(booked_count/total_seats)*100
    if occupancy == 100:
        status="Housefull"
        suggestion="Consider opening an additional show."
    elif occupancy >=75:
        status="Almost Full"
        suggestion="Monitor bookins closely."
    elif occupancy>=50:
        status="Filling Fast"
        suggestion="Seats are available."

    print(f"Occupancy:{occupancy:.0f}%")
    print(f"Show Status:{status}")
    print(f"Suggestion:{suggestion}")

total_seats=200
booked_seats=[1]*150

seat_booking(total_seats,booked_seats)
        
```

```python
def server_load(cpu_readings):
    total=0
    for reading in cpu_readings:
        total+=reading

    average=total/len(cpu_readings)

    if average<50:
        status="Normal"
    elif 50<= average<=80:
        status="Warning"
    else:
        status="Critical"
    print(f"Average Cpu Load:{average:.0f}%")
    print(f"Server Status:{status}")

cpu_readings=[45,60,70,85,90]
server_load(cpu_readings)
```

```python
def classroom_resources(resource_usage):

    thresholds={
        "Projector":5,
        "AC":8,
        "Lights":6
    }

    overused=[]
    for resource, hours in resource_usage.items():
        if resource in thresholds and hours > thresholds[resource]:
            overused.append(resource)
    energy_alert="Yes" if overused else "No"

    print("Overused Resources:", ", ".join(overused) if overused else "None")
    print("Energy Alert:", energy_alert)

resource_usage={
    "Projector":6,
    "AC":9,
    "Lights":4
 }
classroom_resources(resource_usage)
```

```python
def event_registrations_control(event_capacity,registrations):
    confirmed=0
    waitlisted=0

    for i in range(1,registrations+1):
        if confirmed < event_capacity:
            confirmed+=1
        else:
            waitlisted +=1

    status="Closed" if confirmed == event_capacity else "open"

    print("confirmed Registerations:",confirmed)
    print("Waitlisted users:", waitlisted)
    print("Registrations Status:", status)

event_capacity=100
registrations=105

event_registrations_control(event_capacity,registrations)
```
