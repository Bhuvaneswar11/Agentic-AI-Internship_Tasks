```python
sentence="python is easy and python is powerful"   

words=sentence.split()
uniq_words=[]

for word in words:
    if word not in uniq_words:
        uniq_words.append(word)

print("unique words count:",len(uniq_words))
print("unique words:",uniq_words)
```

```python
employees={"Ravi":75000,
          "Anita":68000,
          "Kiran":72000
          }
highest_one=""
highest_salary=0

for name,salary in employees.items():
    if salary>highest_salary:
        highest_salary=salary
        high_one=name

print("highest salary:",high_one,"-",highest_salary)
```

```python
numbers=[45,22,89,10,66]

maximum=numbers[0]
minimum=numbers[0]

for num in numbers:
    if num>maximum:
        maximum=num
    if num<minimum:
        minimum=num
print("List:",numbers)
print("max:",maximum)
print("min:",minimum)
```

```python
prices=[450,1200,899,1500,300]

Pro_above=0

for price in prices:
    if price>=1000:
        Pro_above+=1
print("Products above 1000:",Pro_above)
```

```python
attendance=["P", "P", "A", "P", "P"]

present_days=attendance.count("P")
total_days=len(attendance)

attendance_percentage=(present_days/total_days)*100

print("Attendance Percentage:",attendance_percentage)
```

```python
text="pythonp"

char_freq={}

for char in text:
    if char in char_freq:
        char_freq[char]+=1
    else:
        char_freq[char]=1
        
print(char_freq)
```
