```python
Student_marks=[45,78,90,33,60]

pass_count=0
fail_count=0

for marks in Student_marks:
    if marks>50:
        pass_count+=1
    else:
        fail_count+=1

print("Total Students:",len(Student_marks))
print("Number of Students Passed:",pass_count)
print("Number of Students Failed:",fail_count)

        
```

```python
Names = [" Alice ", "bob", " CHARLIE "]

clean_names=[]

for name in Names:
    clean_name=name.strip().lower()
    clean_names.append(clean_name)

print("Original names:",Names)
print("Cleaned_names:",clean_names)



       
```

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    length=len(message)
    print("Message:",message)
    print("Length:",length)

    if length>10:
        print("status:long message(more than 10 characters)")
    else:
        print("status:short message")

    print()
    
```

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count=0

for log in logs:
    if log == "ERROR":
        error_count+=1
print("Total Error entries:",error_count)
```
