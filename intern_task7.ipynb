```python
class SmartLight:
    def __init__(self,name):
        self.name=name
        self.status="OFF"

    def turn_on(self):
        self.status="ON"
    def turn_off(self):
        self.status="OFF"
    def display_status(self):
        print(f"{self.name} is {self.status}")

light=SmartLight("Bedroom Light")
light.turn_on()

light.display_status()
```

```python
class Employee:
    def __init__(self,name,emp_id,department):
        self.name=name
        self.emp_id=emp_id
        self.department=department

    def display_id_card(self):
        print("Employee ID Card ")
        print(f"Name:{self.name}")
        print(f"ID:{self.emp_id}")
        print(f"Department:{self.department}")
       

employee =Employee("Rahul","EMP102","IT")
employee.display_id_card()
        
```

```python
class Store:
    def __init__(self,contact_name,phone_number):
        self.contact_name= contact_name
        self.phone_number= phone_number

    def contact_saved(self):
        print("Contact Saved")
        print(f"Contact Name:{self.contact_name}")
        print(f"Phone Number:{self.phone_number}")

Store_Contact=Store("Anita","9876543210")
Store_Contact.contact_saved()
```

```python
class stores:
    def __init__(self,product_name,price):
     self.product_name=product_name
     self.price=price

    def format_price_tag(self):
        print(f"Product:{self.product_name}")
        print(f"Price:{self.price}")

price_tag=stores("Headphones","₹2499")
price_tag.format_price_tag()
```

```python
class movie:
    def __init__(self,name,rating):
        self.name=name
        self.ratings=rating

    def display_movie_details(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.ratings}/5")

movie_details=movie("Inception",4.8)
movie_details.display_movie_details()
```

```python
class shop:
    def __init__(self,customer_name,delivery_address):
        self.customer_name=customer_name
        self.delivery_address=delivery_address

    def delivery_details(self):
        print("Delivery Details")
        print(f"Customer:{self.customer_name}")
        print(f"Address:{self.delivery_address}")

store_data=shop("Suman","Hyderabad")
store_data.delivery_details()
```
