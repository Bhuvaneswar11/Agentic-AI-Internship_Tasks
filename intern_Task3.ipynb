```python
employees={
    "Ravi":92,
    "Anita":88,
    "Kiran":92,
    "Suresh":85
}
high_score=0
for name in employees:
    if employees[name]>high_score:
        high_score=employees[name]

top_performers=[]
for name in employees:
    if employees[name] == high_score:
        top_performers.append(name)


        
print(f"Top Performers Eligible for Bonus:{', '.join(top_performers)} (Score: {high_score})")
```

```python
sentence="Buy mobile phone buy phone online"

sentence=sentence.lower()

import string
for p in string.punctuation:
    sentence=sentence.replace(p, "")

words=sentence.split()

frequency={}
for word in words:
    if word in frequency:
        frequency[word]+=1
    else:
        frequency[word]=1
result={word: count for word, count in frequency.items() if count>1}

print(result)

```

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings=[]

for hour, reading in enumerate(sensor_readings):
    if reading % 2 == 0:
        valid_readings.append((hour, reading))
print("Valid Sensor Readings (Hour,Value):")
print(valid_readings)


```

```python
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]
domain_counts = {}


for mail in emails:
    domain=mail.split("@")[1]

    if domain in domain_counts:
        domain_counts[domain]+=1
    else:
        domain_counts[domain]=1
  
total=len(emails)
domain_percentages={}
for domain,count in domain_counts.items():
    domain_percentages[domain]=int((count/total)*100)

for domain,percentage in domain_percentages.items():
    print(f"{domain}: {percentage}%")

```

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

avg_sales=sum(sales)/len(sales)

threshold=avg_sales * 1.3 

for i, value in enumerate(sales, start=1):
    if value >= threshold-20: #tolerance for rounding
         print(f"Day {i}: {value}")
```
