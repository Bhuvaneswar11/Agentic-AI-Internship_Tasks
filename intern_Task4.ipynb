```python
def mobile_recharge(amount):

    if amount < 50:
        print(" Recharge amount must be alteast $50")
        return False
    valid_plans=[199,239,299,399,666]
    if amount in valid_plans:
        print(f"Recharge Successful with plan ${amount},")
        return True
    else:
        print("Inavlid plan. Valid plans are : 199,239,299,399,666.")
        return False

while True:
    try:
        user_input=int(input("Enter recharge amount:$"))
        if mobile_recharge(user_input):
            break
        else:
            print("Please try again.\n")
    except VlaueError:
        print("Invalid input.Please enter a number.\n")
```

```python
def check_inventory(stock_dict):
    for product,quantity in stock_dict.items():
        if quantity<15:
            print(f"{product}:ReOrder Alert (Stock={quantity})")
        else:
            print(f"{product}:Stock OK (Stock ={quantity})")

inventory={
    "Apples":10,
    "Bananas":25,
    "Oranges":14,
    "Mangoes":30,
    "Grapes":5
}
check_inventory(inventory)
```

```python
def process_result(marks):
    total=0
    for mark in marks:
        total+= mark
    average = total/len(marks)

    if average>=50:
        return f"Pass(Average={average:.2f})"
    else:
        return f"Fail (Average={average:.2f})"

Student_marks=[60,45,70,55,40]
result=process_result(Student_marks)
print(result)
 
```

```python
def calculate_fare(distance,peak_hour=False):
    base_fare=50
    per_km_rate=12
    fare=base_fare+(distance*per_km_rate)

    if peak_hour:
        fare*=1.25
    return fare
while True:
    km=float(input("Enter distance travelled (in km): "))
    peak=input("Is it peak hour?(yes/no):").lower()

    total_fare=calculate_fare(km, peak =="yes")
    print(f"Estimated Fare:${total_fare:.2f}")

    retry=input("Do you want to calculate again?(yes/no):").lower()
    if retry!= "yes":
        print("Thank You for using Cab Fare Estimator!")
        break
```

```python
def employee_attendance(attendance_list):
    present_days=0

    for day in attendance_list:
        if day =="P":
            present_days+=1
            
    total_days=len(attendance_list)
    attendance_percentage=(present_days/total_days)*100

    if attendance_percentage >=75:
         return f"Eligible (Attendance={attendance_percentage:.2f}%)"
    else:
        return f"Not Eligible (Attendence={attendance_percentage:.2f}%)"

attendance=["P","A","P","P","A","P","P","P","A","P"]
result=employee_attendance(attendance)
print(result)
```

```python
def check_pswd_strngth(password):
  if len(password)<8:
      return "Weak password(Too Short)"

  has_digit=False
  has_special=False

  for char in password:
         if char.isdigit():
             has_digit=True
         if char in "@#$":
             has_special=True

  if has_digit and has_special:
        return "Strong Password"
  else:
        return "Weak Password(Missing digit or special character)"
print(check_pswd_strngth("ajed124"))
print(check_pswd_strngth("afijed12"))
print(check_pswd_strngth("ajhd@124"))
```
